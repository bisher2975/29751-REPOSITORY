```python
def Core(datasets, threshold):
    """
    Identify highly correlated columns in a dataset.

    Args:
        datasets (pd.DataFrame): The input dataset.
        threshold (float): The correlation threshold.

    Returns:
        set: A set of column names that are highly correlated.
    """
    col_corr = set()  # Using set to store all the names of correlated columns
    corr_matrix = datasets.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:  # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr # Corrected indentation for the return statement
```

```python
corr_features = Core(X_train, 0.7)  # Call the function and assign the result
print(len(set(corr_features)), corr_features)  # Print the desired output
```
